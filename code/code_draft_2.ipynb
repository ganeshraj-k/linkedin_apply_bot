```python
def date_filter(date_posted, val):
    date_options = date_posted.find_elements( "xpath" , ".//li[@class = 'search-reusables__collection-values-item'] ")
    date_posted.click()
    label = date_options[val].find_element("xpath" , ".//label")
    label.click()
    results = date_posted.find_element("xpath" , ".//button[@class = 'artdeco-button artdeco-button--2 artdeco-button--primary ember-view ml2'] ")
    results.click()

def exp_filter( experience, val):
    exp_options = experience.find_elements("xpath" , ".//li[@class = 'search-reusables__collection-values-item'] ")
    experience.click()
    label = exp_options[val].find_element("xpath" , ".//label")
    label.click()
    results = experience.find_element("xpath" , ".//button[@class = 'artdeco-button artdeco-button--2 artdeco-button--primary ember-view ml2'] ")
    results.click()

def salary_filter( salary, val):
    salary_options = salary.find_elements("xpath" , ".//li[@class = 'search-reusables__collection-values-item'] ")
    salary.click()
    label = salary_options[val].find_element("xpath" , ".//label")
    label.click()
    results = salary.find_element("xpath" , ".//button[@class = 'artdeco-button artdeco-button--2 artdeco-button--primary ember-view ml2'] ")
    results.click()


def hybrid_filter( hybrid, val):
    hybrid_options = hybrid.find_elements("xpath" , ".//li[@class = 'search-reusables__collection-values-item'] ")
    hybrid.click()
    label = hybrid_options[val].find_element("xpath" , ".//label")
    label.click()
    results = hybrid.find_element("xpath" , ".//button[@class = 'artdeco-button artdeco-button--2 artdeco-button--primary ember-view ml2'] ")
    results.click()

def easy_apply_filter():
    easy_apply_button = driver.find_element( "xpath" , ".//button[contains(@aria-label ,'Easy Apply filter')]") 
    easy_apply_button.click()
```

```python
def apply_job(job_card):
    job_card.click()
    easy_apply_button = driver.find_element("xpath", "//button[contains(@aria-label , 'Easy Apply to')]")
    easy_apply_button.click()

    
    isform = True
    fillresume = False
    
    while isform:

        if theresubmit(driver):
            find_submit(driver)
            break

        try:
            apply = driver.find_element("xpath" , ".//div[@class='jobs-easy-apply-content']")
            form = apply.find_element("xpath", ".//form")
            print("form found")
        except Exception as f:
            print("form not found ".format(f))
            isform = False
        
        
        

        if not fillresume:
            print("resume" )
            if there_resume(form):
                print("fill resume")

                resumes = form.find_elements("xpath" ,".//div[contains(@class, 'ui-attachment jobs-document-upload-redesign-card__container')]")
                print(len(resumes))
                driver.execute_script("arguments[0].click();", resumes[0]) 
                resumes[0].click()
                # enter_resume(form, 0)
                fillresume = True
        
        fields = form.find_elements("xpath" , "//div[contains(@class, 'jobs-easy-apply-form-section__grouping')]")

        for field in fields:
            print(QnA(field.text.lower()))
            print(buttontype(field))
            if buttontype(field) == 'multi_select' :
                answer_select( field, QnA(field.text.lower()))
            elif buttontype(field) == 'text':
                answer_text(field, QnA(field.text.lower()))
            elif buttontype(field) == 'radio' :
                answer_radio(field, QnA(field.text.lower()))

        find_next(form)

        driver.implicitly_wait(2)
 


    driver.implicitly_wait(5)

    # done_buttons  =  driver.find_elements("xpath" , ".//button[@aria-label = 'Dismiss']")
    # done_buttons  =  driver.find_elements("xpath" , ".//button[@aria-label = 'Dismiss']")
    # done_button  =  driver.find_element("xpath" , ".//button[@aria-label = 'Dismiss']")
    # try:
    #     done_button.click()
    # except:
    #     done_button  =  driver.find_element("xpath" , ".//button[@aria-label = 'Dismiss']")
    #     done_button.click()
    submit_button = driver.find_element("xpath" , ".//button[@class = 'artdeco-button artdeco-button--2 artdeco-button--primary ember-view mlA block' and .//span[text()='Done' ]]")
    submit_button.click()
    
```
