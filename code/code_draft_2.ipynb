```python
def date_filter(date_posted, val):
    date_options = date_posted.find_elements( "xpath" , ".//li[@class = 'search-reusables__collection-values-item'] ")
    date_posted.click()
    label = date_options[val].find_element("xpath" , ".//label")
    label.click()
    results = date_posted.find_element("xpath" , ".//button[@class = 'artdeco-button artdeco-button--2 artdeco-button--primary ember-view ml2'] ")
    results.click()

def exp_filter( experience, val):
    exp_options = experience.find_elements("xpath" , ".//li[@class = 'search-reusables__collection-values-item'] ")
    experience.click()
    label = exp_options[val].find_element("xpath" , ".//label")
    label.click()
    results = experience.find_element("xpath" , ".//button[@class = 'artdeco-button artdeco-button--2 artdeco-button--primary ember-view ml2'] ")
    results.click()

def salary_filter( salary, val):
    salary_options = salary.find_elements("xpath" , ".//li[@class = 'search-reusables__collection-values-item'] ")
    salary.click()
    label = salary_options[val].find_element("xpath" , ".//label")
    label.click()
    results = salary.find_element("xpath" , ".//button[@class = 'artdeco-button artdeco-button--2 artdeco-button--primary ember-view ml2'] ")
    results.click()


def hybrid_filter( hybrid, val):
    hybrid_options = hybrid.find_elements("xpath" , ".//li[@class = 'search-reusables__collection-values-item'] ")
    hybrid.click()
    label = hybrid_options[val].find_element("xpath" , ".//label")
    label.click()
    results = hybrid.find_element("xpath" , ".//button[@class = 'artdeco-button artdeco-button--2 artdeco-button--primary ember-view ml2'] ")
    results.click()

def easy_apply_filter():
    easy_apply_button = driver.find_element( "xpath" , ".//button[contains(@aria-label ,'Easy Apply filter')]") 
    easy_apply_button.click()
```

```python
def apply_job(job_card):
    role_name = job_card.find_element( "xpath" , ".//div[@class = 'full-width artdeco-entity-lockup__title ember-view']")
    company_name = job_card.find_element( "xpath" , ".//div[@class = 'artdeco-entity-lockup__subtitle ember-view']")
    other_data = job_card.find_elements("xpath" , ".//ul[@class = 'job-card-container__metadata-wrapper']")

    data_tuple = [role_name.text.split('\n')[0], company_name.text, [ other.text for other in other_data]]
    
    job_card.click()

    try:
        easy_apply_button = driver.find_element("xpath", ".//button[contains(@aria-label , 'Easy Apply to')]")
        easy_apply_button.click()
        print("clicked easy apply")

    except Exception as e:
        logger.exception("ALREADY APPLIED TO THIS ROLE")
        return
        

    
    isform = True
    fillresume = False
    
    while isform:

        if theresubmit(driver):
            get_submit(driver)
            entries.append(data_tuple)
            with open('applications.txt', 'a') as file:
                file.write(' , '.join(data_tuple) + '\n')
            break

        try:
            apply = driver.find_element("xpath" , ".//div[@class='jobs-easy-apply-content']")
        except Exception as e:
            logger.exception("EASY APPLY SECTION MISSING")
            
        try:
            form = apply.find_element("xpath", ".//form")
        except Exception as e:
            logger.exception("FORM MISSING")
            isform = False
        
        
        

        if not fillresume:
            
            if there_resume(form):
                resumes = form.find_elements("xpath" ,".//div[contains(@class, 'ui-attachment jobs-document-upload-redesign-card__container')]")
                driver.execute_script("arguments[0].click();", resumes[0]) 
                resumes[0].click()
                # enter_resume(form, 0)
                fillresume = True

        try:  
            fields = form.find_elements("xpath" , "//div[contains(@class, 'jobs-easy-apply-form-section__grouping')]")
        except Exception as e:
            logger.exception("FORM MISSING/ FORM NOT LOADED/ FIELDS NOT LOADED")

        for field in fields:
            question = field.text.lower().split('\n')[0]
            if question not in questions:
                questions.append(question)
                with open('questions.txt', 'a') as file:
                    file.write(question + '\n')
            answer = QnA(question)

            if buttontype(field) == 'multi_select' :
                answer_select( field, answer)
            elif buttontype(field) == 'text':
                answer_text(field, answer)
            elif buttontype(field) == 'radio' :
                answer_radio(field, answer)

        find_next(form)

        
 


    try:
        exit_button = driver.find_element("xpath" , ".//button[@class = 'artdeco-button artdeco-button--2 artdeco-button--primary ember-view mlA block' and .//span[text()='Done' ]]")
        exit_button.click()
    except Exception as e:
        logger.exception("FAILED TO EXIT APPLICATION SUBMITTED POP-UP")
    
```

```python
def iterator( driver):
    currentpage = 1
    nextpageexists = True
    
    while nextpageexists:
        try:
            search_results = driver.find_element("xpath" , ".//div[contains(@class, 'jobs-search-results-list')]")
            list_of_jobs = search_results.find_elements( "xpath" , "//li[@data-occludable-job-id]")
        except Exception as e:
            logger.exception("FAILED TO RETRIEVE SEARCH RESULTS")
            return
        
        driver.execute_script("arguments[0].scrollIntoView();", list_of_jobs[-1])
        for job in list_of_jobs:
            job.click()
            apply_job(job)
            
        
        try:
            next_page = driver.find_element("xpath" , "//li[@data-test-pagination-page-btn = {}]".format(currentpage+1))
            next_page.click()
            currentpage+=1
        except Exception as e:
            logger.exception("END OF RESULTS / CANNOT FIND NEXT PAGE")
            nextpageexists = False
            
```
